```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
from datetime import datetime
```

```python
from pathlib import Path

path = Path('C:\\')/'Users'/'Lenovo'/'Desktop'/'python anaconda'/'covid.csv'
data = pd.read_csv(path)
```

```python
th = np.where(data == 'THA')

cases = np.array(data['cases'][th[0][0]:th[0][-1] + 1])
```

```python
print('COVID-19 in Thailand')
print('Cases :',np.sum(cases))
print("Percentage of infected people in the entire population :", float("{:.4f}".format(np.sum(cases)/69428524*100)),'%')
print('Deaths :', np.sum(data['deaths'][th[0][0]:th[0][-1] + 1]))
print('Percentage of deat per case :', float("{:.4f}".format(np.sum(data['deaths'][th[0][0]:th[0][-1] + 1])/np.sum(cases)*100)),'%')
```

```python
def dataplot(code):
    country = np.array(np.where(data == code))
    cases = np.array(data['cases'][country[0][0]:country[0][-1]+1])
    day_list = data['dateRep'][country[0][0]:country[0][-1]+1] 
    day = [datetime.strptime(x,'%d/%m/%Y' ) for x in day_list]

    sum = [np.sum(cases),]
    s = np.sum(cases)
    for i in range(len(day_list)-1):
        s = s-cases[i]
        sum.append(s)
    return day,sum
```

```python
country_list = [('THA','Thailand','*'), ('TWN', 'Taiwan', '+'), ('SYR', 'Syria', 'x'),\
                ('CHE', 'Switzerland', 'X'), ('SWE', 'Sweden', 'D'), ('SUR','Suriname', '^'),\
                ('SDN','Sudan', 'v'), ('SSD', 'South Sudan', 's'), ('LKA', 'Sri Lanka', '1'),\
                ('ESP', 'Spain', '4')] 
```

```python
figure(num=None, figsize=(10, 10), dpi=500, facecolor='w', edgecolor='k')

for code, country, symb in country_list:
  x,y = dataplot(code)
  plt.plot_date(x,y,symb, label=country)

plt.yscale('log')

plt.xticks(rotation=45,fontsize=16)
plt.yticks(fontsize=16)
plt.legend(loc='center left',fontsize=16, bbox_to_anchor=(1, 0.5))
plt.title('COVID-19 Infection Rate since 31/12/2019 to 12/04/2020',fontsize=20, fontweight='bold')
plt.xlabel('Date',fontsize=20)
plt.ylabel('Cases',fontsize=20)
plt.grid(True, which="both", ls="--",alpha=0.5)
plt.show()
```

```python
def dataplot2(code):
    country = np.array(np.where(data == code))
    cases = np.sum(np.array(data['cases'][country[0][0]:country[0][-1]+1]))
    deaths = np.sum(np.array(data['deaths'][country[0][0]:country[0][-1]+1]))
    population = data['popData2018'][country[0][0]]

    cases_per_population = cases/population*100
    death_per_cases = deaths/cases*100

    population = np.sqrt(population)/50
    
    return cases_per_population,death_per_cases,population
```

```python
figure(num=None, figsize=(10, 10), dpi=500, facecolor='w', edgecolor='k')
plt.set_cmap('hot')

for code, country, symb in country_list:
  x,y,z = dataplot2(code)
  plt.scatter(x,y,z, label=country,alpha=0.5)


plt.legend(loc='upper left',fontsize=16, bbox_to_anchor=(1, 0.5))
plt.title('COVID-19 cases and deaths \n from 31-12-2019 to 12-04-2020',fontsize=20, fontweight='bold')
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.xlabel('Cases per population (%)',fontsize=20)
plt.ylabel('Death per cases (%)',fontsize=20)
plt.grid(True, which="both", ls="--")

plt.show()
```
